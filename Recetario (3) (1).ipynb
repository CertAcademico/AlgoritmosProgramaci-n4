# Sientes que falta algun termino? Agregalo tu mismo! 

```python
from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets
## Estos son las librerias que usaremos para general un espacio donde el usuario sea capaz de aadir sus propias entradas

def aadir_entrada(sender):
    Nombre = Nombre_input.value
    Termino = Termino_input.value
    Significado = Significado_input.value

    Texto_pantalla = f"##{Termino}\n**Nombre:** {Nombre}\n{Significado}"
## Con este fracmento de codigo se muestra en pantalla el orden en que el texto saldria

    clear_output()
    display(Markdown(Texto_pantalla))

Nombre_input = widgets.Text(placeholder = "Ingrese su nombre")
Termino_input = widgets.Text(placeholder = "Ingrese el nombre del termino que desea ingresar")
Significado_input = widgets.Textarea(placeholder = "Ingrese el significado de su termino")

subido = widgets.Button( descripcion = "Subir su termino")

## Con esto permitimos al codigo mostrar en pantalla todos los campos de entrada del que el usuario hara uso.

subido.on_click(aadir_entrada)

display(Nombre_input, Termino_input, Significado_input, subido)
```
